```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([])
y2 = np.array([])
x = np.array([])
with open("yawData.txt") as file_in:
    i = 0
    for line in file_in:
        ar = line.split(' ')
        nav = ar[1].split(':')[1]
        bno = ar[2].split(':')[1]
        y1 = np.append(y1, nav)
        y2 = np.append(y2, bno)
        x = np.append(x,i)
        i += 1



plt.plot(x, y1)
plt.plot(x, y2)

#plt.ylim([-90, 90])
#plt.yticks(np.arange(-90, 90, 10))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.minorticks_off()
ticks = np.arange(0, 90, 10)
plt.set_yticks(ticks)
plt.grid(linestyle='--')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(0, 10, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(y, color='blue', label='Sine wave')
ax.plot(z, color='black', label='Cosine wave')

plt.show()
```
